```python
import requests
from bs4 import BeautifulSoup

def scrape_internships(url, keyword="data science"):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "lxml")

    jobs = []
    for job_card in soup.find_all("div", class_="job_seen_beacon"):
        title = job_card.find("h2").text.strip() if job_card.find("h2") else "No title"
        company = job_card.find("span", class_="companyName").text.strip() if job_card.find("span", class_="companyName") else "Unknown"
        link = job_card.find("a")["href"] if job_card.find("a") else None

        if keyword.lower() in title.lower():
            jobs.append({"title": title, "company": company, "link": link})

    return jobs

# Example: scraping Indeed (adjust URL for your region)
url = "https://pk.indeed.com/jobs?q=data+science+internship&l=Pakistan"
internships = scrape_internships(url)

for job in internships[:5]:
    print(job)
```

```python
def draft_email(job):
    return f"""
Dear {job['company']} Recruitment Team,

I am writing to apply for the {job['title']} position at {job['company']}.
As a Data Science student with strong skills in Python, scikit-learn, and professional workflow documentation,
I am eager to contribute to your team and grow through this internship opportunity.

Please find my resume and portfolio attached. I look forward to your response.

Best regards,
Umair Khitab
"""

for job in internships[:3]:
    print(draft_email(job))
```
