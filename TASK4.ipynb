```python
import re
from PyPDF2 import PdfReader # Import PdfReader again as it's needed here

search_term = "h"

# Re-open the PDF file and extract all text to create full_text
file_name = "/content/Wiley_The_Data_Warehouse_Toolkit_The_Definitive_Guide_to_Dimensional.pdf" # Use the correct file path
full_text = ""
with open(file_name, "rb") as file:
    reader = PdfReader(file)
    for page in reader.pages:
        full_text += page.extract_text() if page.extract_text() else ""

matches = re.findall(search_term, full_text, re.IGNORECASE)

if matches:
    print(f"Found '{search_term}' {len(matches)} times in the book.")

else:
    print(f"'{search_term}' was not found in the book.")
```

# Chapter: Regular Expressions Applied to My Book

## Meta Characters
We use `\d` to find digits, `\w` to find words, and `\s` to find spaces.

## Quantifiers
We use `+` for one or more, `*` for zero or more, and `{n}` for exact counts.

## Escape Codes
We use `\.` to match literal dots and `\\` to match backslashes.


```python
import re

# Example: Extract text from first page
first_page_text = reader.pages[0].extract_text()

# Find all numbers
numbers = re.findall(r"\d+", first_page_text)
print("Numbers found:", numbers[:20])

# Find all email-like patterns
emails = re.findall(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", first_page_text)
print("Emails found:", emails)
```
